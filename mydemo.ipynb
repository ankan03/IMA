### 2. Cleaning the data and getting it ready

```python
data.isnull().sum()
```

```python
data['document_create_date'].equals(data['document_create_date.1'])
```

```python
new_data = data.drop(['posting_id'], axis=1) #whole column is constant
new_data = new_data.drop(['document_create_date'], axis=1) #
new_data.dropna(how='all', axis=1, inplace=True) #whole column is NaN
new_data.head()
```

```python
new_data['clear_date']= pd.to_datetime(new_data['clear_date'])
# new_data['buisness_year']= pd.to_datetime(new_data['buisness_year'],format='%Y')
new_data['posting_date']= pd.to_datetime(new_data['posting_date'])
new_data['document_create_date.1']= pd.to_datetime(new_data['document_create_date.1'],format='%Y%m%d')
new_data['due_in_date']= pd.to_datetime(new_data['due_in_date'],format='%Y%m%d')
new_data['baseline_create_date']= pd.to_datetime(new_data['baseline_create_date'],format='%Y%m%d')
new_data.dtypes
```

```python
new_data['delay']=new_data['clear_date']-new_data['due_in_date']
```

```python
new_data.head()
```

```python
new_data=pd.get_dummies(new_data, columns=['business_code'])
new_data
```

```python
new_data['cust_number']=new_data['cust_number'].astype("category").cat.codes
new_data['name_customer']=new_data['name_customer'].astype("category").cat.codes
new_data['invoice_currency']=new_data['invoice_currency'].astype("category").cat.codes
new_data['document type']=new_data['document type'].astype("category").cat.codes
new_data['cust_payment_terms']=new_data['cust_payment_terms'].astype("category").cat.codes
new_data
```
